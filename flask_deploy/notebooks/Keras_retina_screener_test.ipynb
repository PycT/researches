```python
import tensorflow as tf;
import os, random;
from PIL import Image;
import matplotlib.pyplot as plt;
import numpy as np;

#model_path = "../models/retina_screening.h5";
#model_path = "../models/nasnet_retina_screening.h5";
#model_path = "../models/inceptionv3_retina_screening299.h5";
model_path = "../models/inceptionv3_retina_screening1.h5";

#path_to_images1 = "../../../Datasets/STARE/validation/normal/";
#path_to_images2 = "../../../Datasets/STARE/validation/abnormal/";
path_to_images1 = "../../../Datasets/STARE/training/normal/";
path_to_images2 = "../../../Datasets/STARE/training/abnormal/";

input_shape = (600, 600, 3);

model = tf.keras.models.load_model(model_path);
model.summary();
```

```python
images = [];

counter = 0;
for img in os.listdir(path_to_images1):
    if os.path.isfile(path_to_images1 + img):
        if img !=".DS_Store":
            images.append({"image": path_to_images1+img, "label": "normal"});
            print(images[-1]);
            counter += 1;
    if (counter >= 10):
        break;

counter = 0;
for img in os.listdir(path_to_images2):
    if os.path.isfile(path_to_images2 + img):
        if img !=".DS_Store":
            images.append({"image": path_to_images2+img, "label": "abnormal"});
            print(images[-1]);
            counter += 1;
    if (counter >= 10):
        break;

random.shuffle(images);
```

```python
normalizer = tf.keras.preprocessing.image.ImageDataGenerator\
(
    featurewise_center = True,
    featurewise_std_normalization = True
);
for sample in images:
    img = tf.keras.preprocessing.image.load_img\
    (
        sample["image"],
        target_size = (input_shape[0], input_shape[1])
    );
    
    img_array = tf.keras.preprocessing.image.img_to_array(img);
    #img_shaped = normalizer.flow(img_array.reshape((1,) + img_array.shape), batch_size = 1);
    img_shaped = img_array.reshape((1,) + img_array.shape);
    
    res = model.predict(img_shaped);

    pic = Image.open(sample["image"]);
    plt.imshow(np.asarray(pic));
    plt.show();
    
    inference = res[0][0] * 1000 - 453;
    
    if inference > 0.5:
        prediction = "normal";
    else:
        prediction = "abnormal";
    
    if sample["label"] == prediction:
        prediction += " (correct)";
    else:
        prediction += " (incorrect)";
    print(" Image: {}\n Label: {}\n Prediction: {}\n Inference: {}".format\
          (
            sample["image"], 
            sample["label"], 
            prediction, 
            inference
          ));
    print();
    """
    print(res[0][0]);
    """
```
